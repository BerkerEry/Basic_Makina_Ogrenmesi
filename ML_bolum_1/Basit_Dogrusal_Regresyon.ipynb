```python
import pandas as pd
```

```python
df = pd.read_csv("Advertising.csv")
df = df.iloc[:, 1: len(df)]
df.head()
```

```python
"""
bir şirketin reklam harcamalarının satışlarının ne şekilde değişeceğini görmek isteriz
"""

# basit doğrusal regresyon
```

```python
df.info()
```

```python
import seaborn as sns
sns.jointplot(x = "TV", y = "sales", data = df, kind = "reg")
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = df[["TV"]]
X.head()
```

```python
y = df[["sales"]]
y.head()
```

```python
reg = LinearRegression()
model = reg.fit(X, y) # X bağımsız, y bağımlı değişken
```

```python
model
```

```python
str(model)
```

```python
dir(model)
```

```python
model.intercept_ # B0 katsayisi
```

```python
model.coef_ # B1 katsayisi
```

```python
# rkare
model.score(X, y) # satislardaki degisikligin yaklasik %60 i bagimsiz degiskence aciklanmaktadir
```

## Artıklar ve Makine Öğrenmesindeki Önemi

### MSE: Hata Kareler Ortalaması

### RMSE: Hata Kareler Ortalamasının Karekökü

```python
y.head()
```

```python
model.predict(X)[0:5]
```

```python
gercek_y = y[0:10]
```

```python
tahmin_edilen_y = pd.DataFrame(model.predict(X)[0:10])
```

```python
hatalar = pd.concat([gercek_y, tahmin_edilen_y], axis = 1)
```

```python
hatalar
```

```python
hatalar.columns = ["gercek_y", "tahmin_edilen_y"]
hatalar
```

```python
hatalar["hata"] = hatalar["gercek_y"] - hatalar["tahmin_edilen_y"]
hatalar
```

```python
hatalar["hata_kareler"] = hatalar["hata"]**2
hatalar
```

```python
import numpy as np
np.mean(hatalar["hata_kareler"])
```
