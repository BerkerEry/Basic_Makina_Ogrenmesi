```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale, StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report,mean_squared_error, r2_score, roc_auc_score, roc_curve
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("diabetes.csv")
df.head()
```

```python
y = df["Outcome"]
X = df.drop(["Outcome"], axis = 1)
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=42)
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
X_train = scaler.transform(X_train)
```

```python
scaler.fit(X_test)
X_test = scaler.transform(X_test)
```

## Model ve Tahmin

```python
mlpc_model = MLPClassifier().fit(X_train, y_train)
```

```python
mlpc_model.coefs_
```

```python
y_pred = mlpc_model.predict(X_test)
accuracy_score(y_test, y_pred)
```

## Model Tuning

```python
mlpc_params = {"alpha": [0.01, 0.001,0.0001,0.005],
               "hidden_layer_sizes": [(10,10),(100,100,100),(3,5)],
               }
```

```python
mlpc = MLPClassifier(solver = "lbfgs", activation="logistic")
```

```python
mlpc_cv_model = GridSearchCV(mlpc, mlpc_params, cv=10, n_jobs=-1, verbose=2).fit(X_train, y_train)
```

```python
mlpc_cv_model.best_params_
```

```python
# final model
mlpc_tuned = MLPClassifier(solver="lbfgs", alpha=0.0001, activation="logistic",hidden_layer_sizes=(3,5)).fit(X_train, y_train)
y_pred = mlpc_tuned.predict(X_test)
accuracy_score(y_test, y_pred)
```
