```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import scale, StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report,mean_squared_error, r2_score, roc_auc_score, roc_curve
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier

import warnings
warnings.filterwarnings("ignore")
```

```python
from lightgbm import LGBMClassifier
```

```python
df = pd.read_csv("diabetes.csv")
df.head()
```

```python
y = df["Outcome"]
X = df.drop(["Outcome"], axis = 1)
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=42)
```

## Model ve Tahmin

```python
lgbm_model = LGBMClassifier().fit(X_train, y_train)
```

```python
y_pred = lgbm_model.predict(X_test)
accuracy_score(y_test,y_pred)
```

## Model Tuning

```python
lgbm = LGBMClassifier()
```

```python
lgbm_params = {"learning_rate": [0.001,0.01],
               "max_depth": [2,3]}
```

```python
lgbm_cv_model = GridSearchCV(lgbm, lgbm_params,cv=10, n_jobs=-1,verbose=2).fit(X_train, y_train)
```

```python
lgbm_cv_model.best_params_
```

```python
lgbm_tuned = LGBMClassifier(learning_rate=0.01, max_depth=2).fit(X_train,y_train)
```

```python
y_pred = lgbm_tuned.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
# degisken onem duzeyleri

feature_imp = pd.Series(lgbm_tuned.feature_importances_,
                        index = X_train.columns).sort_values(ascending=False)

sns.barplot(x=feature_imp, y=feature_imp.index)
plt.xlabel("Değişken Önem Skorları")
plt.ylabel("Değişkenler")
plt.title("Değişken Önem DÜzeyleri")
plt.show()
```
